```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
dataset = [['Skirt', 'Sneakers', 'Scarf', 'Pants', 'Hat'],

        ['Sunglasses', 'Skirt', 'Sneakers', 'Pants', 'Hat'],

        ['Dress', 'Sandals', 'Scarf', 'Pants', 'Heels'],

        ['Dress', 'Necklace', 'Earrings', 'Scarf', 'Hat', 'Heels', 'Hat'],

      ['Earrings', 'Skirt', 'Skirt', 'Scarf', 'Shirt', 'Pants']]
```

```python
from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder()
data_encoded = te.fit_transform(dataset)
df_encoded = pd.DataFrame(data_encoded, columns=te.columns_)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
item_frequency = df_encoded.sum().sort_values(ascending=False)
plt.figure(figsize=(8, 6))
item_frequency.plot(kind='bar')
plt.xlabel('Items')
plt.ylabel('Frequency')
plt.title('Item Frequency')
plt.xticks(rotation=45)
plt.show()
```

```python
frequent_itemsets = apriori(df_encoded, min_support=0.1, use_colnames=True)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric='confidence', min_threshold=0.7)
```

```python
print("Frequent Itemsets:")
print(frequent_itemsets)
print("\nAssociation Rules:")
print(rules)
```
